```python
cat_feats = ['job','marital','education','contact','month','poutcome']
```

# Observation:
Categorical varaibles using dummy variables, so sklearn will be able to understand this in one clean step using pd.get_dummies.
Create a list of 6 element containing the string. Call this list cat_feats.    

```python
final_data = pd.get_dummies(bank,columns=cat_feats,drop_first=True)
```

# Observation:
Use pd.get_dummies columns=cat_feats, drop_first to create a fixed larger dataframe that has new feature columns with dummy variables. 
Set this dataframe as final_data.    

```python
final_data.info()
```

# Observation:
Every column considered as numerical value

```python
from sklearn.model_selection import train_test_split
```

# Observation:
Train and Test data split    

```python
X = final_data.drop('loan',axis=1)
y = final_data['loan']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=101)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

# Observation:
Training a Decision Tree Model    

```python
dtree = DecisionTreeClassifier()
```

```python
dtree.fit(X_train,y_train)
```

```python
predictions = dtree.predict(X_test)
```

# Observation:
Predictions and Evaluation of Decision Tree    

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,predictions))
print(confusion_matrix(y_test,predictions))
```

```python
accuracy = (9636+589)/(9636+1740+1599+589)
print(accuracy)
```

# Conclusion:
In the model using Decision Tree accuracy is 75%

```python
from sklearn.ensemble import RandomForestClassifier
```

# Observation:
Training the Random Forest model    

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier(n_estimators=600)
rfc.fit(X_train,y_train)
```

```python
predictions = rfc.predict(X_test)
```

# Observation:
Predictions and Evaluation    

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,predictions))
```

```python
print(confusion_matrix(y_test,predictions))
```

```python
accuracy = (11224+178)/(11224+2010+152+178)
print(accuracy)
```

# Conclusion:
In the model using Random Forest accuracy is 84%
